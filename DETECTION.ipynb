```python
#VISUALIZACION DE MATRIZ DE CORRELACION 
plt.figure(figsize=(12,8))
correlation_matrix=data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
# Normalizar las características (excluyendo la columna 'Class', que es la etiqueta)
scaler = StandardScaler()
data_scaled = data.copy()
data_scaled.iloc[:, :-1] = scaler.fit_transform(data_scaled.iloc[:, :-1])
```

```python
from sklearn.model_selection import train_test_split

#separamos caracteristicas
X=data_scaled.drop("Class",axis=1)
y=data_scaled["Class"]

X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3,random_state=42)

```

```python
from sklearn.ensemble import RandomForestClassifier
#creamos el modelo 

model=RandomForestClassifier(  n_estimators=100,random_state=42)

#entrenanr el modelo 

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
#predicir

y_pred=model.predict(X_test)

#mostramos el reporte de clasificacion

print(classification_report(y_test,y_pred))

#mostramos resultados
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt="d")

plt.title("matriz de confusion")
plt.show()
```
